```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 学习率取值

## 均匀取值

```python
learning_rate = np.random.uniform(1e-6, 1e-1, 1000)
```

```python
sns.kdeplot(learning_rate)
plt.plot([1e-6, 1e-6], [0, 11], c="red")
plt.plot([1e-1, 1e-1], [0, 11], c="red")
```

## log取值,才能保证大量的值取在较小的数

```python
log_learning_rate = np.random.uniform(np.log10(1e-6), np.log10(1e-1), 1000)
```

```python
sns.kdeplot(log_learning_rate)
```

```python
# 将对数取指数还原回去
sns.kdeplot(np.power(10, log_learning_rate))
plt.plot([1e-6, 1e-6], [0, 80], c="red")
plt.plot([1e-1, 1e-1], [0, 80], c="red")
```
